#plot
- bar for value wise comparison
- pie for percent based comparision
- area or line for time series comparision
- scatter for comparing 2 or 3 items in data
- boxplot for understanding the statistic of data
- histogram for understanding the frequency of distribution

```python
df.head()[years].T.plot(kind='area',stacked=False,alpha=.3,figsize=(20,5))
```

```python
df.head()[years].T.plot(kind='hist',alpha=.3,figsize=(5,20),subplots=True)
plt.show()
```
