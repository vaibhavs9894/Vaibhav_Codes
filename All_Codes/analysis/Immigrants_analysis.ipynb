```python

df.rename({'OdName':'Country',
           'AreaName':'Continent',
           'RegName':'Region',
           'DevName':'Status'},axis=1, inplace=True)
df.head()
```

```python
df.set_index('Country',inplace=True)
```

```python
df.head()[2013].plot(kind='bar')
```

```python
cols = ['Type','Coverage','AREA','REG','DEV']
df.drop(cols,axis=1,inplace=True)
df.head()
```

```python

df.groupby('Continent')[1985].sum().plot(kind='pie',figsize=(8,8),title='continent wise immigration in 1985',autopct='%.3f%%')
plt.show()
```

```python
df.groupby('Continent')[2010].sum().plot(kind='pie',figsize=(8,8),title='continent wise immigration in 2010',autopct='%.3f%%')
plt.show()
```
